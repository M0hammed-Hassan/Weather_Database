```python
import sqlite3
import pandas as pd
```

```python
#This is our csv file
df=pd.read_csv('Measurements.csv')
df
```

```python
#Data frame information 
df.info()
```

```python
#At the begining i was created a database named 'Weather' by using command prompt commands
#Connect to database
connection=sqlite3.connect('Weather.db')
#Define cursor function
cursor=connection.cursor()
#I will create a database table named WeatherMeasurements 
cursor.execute('''
                CREATE TABLE WeatherMeasurements
                (Temp real,
                Wind real,
                Humidity real,
                Visibility real,
                Pressure real,
                Time_Now datetime)
                ''')
```

```python
#Inserting our data into database table
for i in range(len(df)):
    cursor.execute('''INSERT INTO WeatherMeasurements VALUES(?,?,?,?,?,?)''',[df['Temp(C)'][i],df['Wind(km/h)'][i],df['Humidity(%)'][i],df['Visibility(m)'][i],df['Pressure(mb)'][i],df['Time_Now(D/M/Y H:M:S)'][i]])
    #Making a commit
    connection.commit()
#Close our database connection
```

```python
#Displaying database data
connection=sqlite3.connect('Weather.db')
data=pd.read_sql('SELECT * FROM WeatherMeasurements',connection)
data
```
